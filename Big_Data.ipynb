### Encoder

```python
from sklearn.preprocessing  import OrdinalEncoder, LabelEncoder
```

```python
le = LabelEncoder()
df['Item_Visibility'] = le.fit_transform( df[["Item_Visibility"]])
```

### Select all the categorical columns

```python
catcol = df.select_dtypes(object).columns
```

```python
oe = OrdinalEncoder()

df[ catcol ] = oe.fit_transform(df[ catcol ])
```

### Divide the data into features and response

```python
x = df.iloc[ : , :-1] # 2D
y = df.iloc[ : ,  -1] # 1D
```

### Train & test

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split( x, y,  test_size=0.2, random_state=1)
```

### model section

```python
from sklearn.linear_model import  LinearRegression

linreg = LinearRegression()

linreg.fit(xtrain, ytrain)

yp = linreg.predict(xtest)
```

```python
yp
```

### model evalution

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

print("R2 Score :", r2_score(ytest, yp))
print("MAE :", mean_absolute_error(ytest, yp))
print("MSE :", mean_squared_error(ytest, yp))
print("RMSE :", np.sqrt(mean_squared_error(ytest, yp)))
```
